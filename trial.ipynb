```python
from deepface import DeepFace as DF
```

```python
models = [
  "VGG-Face",
  "Facenet",
  "Facenet512",
  "OpenFace",
  "DeepFace",
  "DeepID",
  "ArcFace",
  "SFace",
]
```

```python
for i in range(len(models)):
    verification = DF.verify("ak.jpg", "img1.jpeg", detector_backend = 'dlib', model_name = models[i])

    print(verification)
```
